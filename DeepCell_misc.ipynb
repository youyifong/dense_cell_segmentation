**Explore Datasets in deepcell.datasets**<br>
None of these datasets that can be loaded through deepcell.datasets.xx.load_data is of 512x512.

```python
import deepcell
from skimage import io
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
(X_train, y_train), (X_test, y_test) = deepcell.datasets.hek293.load_data()
print(X_train.shape)
print(X_test.shape)
```

```python
(X_train, y_train), (X_test, y_test) = deepcell.datasets.hela_s3.load_data()
print(X_train.shape)
print(X_test.shape)
```

```python
(X_train, y_train), (X_test, y_test) = deepcell.datasets.nih_3t3.load_data()
print(X_train.shape)
print(X_test.shape)
```
